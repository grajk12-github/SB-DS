```python
import pandas_datareader.data as web
import datetime
import pandas as pd
```

```python
start = datetime.datetime(2013, 1, 1)   # histroy data load range
end = datetime.date.today()
msft = web.DataReader("MSFT",'yahoo',start, end)
aapl = web.DataReader("AAPL",'yahoo',start, end)
```

```python
aapl
```

```python
# Cocatanation and Pivoting and add Symbal to the respective dataframes
msft.insert(0,'Symbol','MSFT')
aapl.insert(0,'Symbol','AAPL')
madf = pd.concat([msft, aapl])
```

```python
madf.sort_index() # sort the index after concatenation
```

```python
#Pivot on date extract AdjClose for all stocks/Symbol, sort the index and reset the index values as the merge cause index mismatch
madf = madf.sort_index()
madf = madf.reset_index() #reset the index
madf.pivot(index='Date', columns='Symbol',values='Adj Close')
```

# split-apply-combine pattern.

```python
madf.insert(1, 'Year', pd.DatetimeIndex(madf['Date']).year) # add the year column and populate
```

```python
madf.insert(2, 'Month',pd.DatetimeIndex(madf['Date']).month) # add the month column and populate
```

```python
madf.groupby(['Symbol','Year','Month']).mean()
```

Time-series data and modeling 
Define a function to get the list of Sympols/Tickers into a single dataframe 
Apple (AAPL), Microsoft (MSFT), General Electric (GE), Google (GOOG), IBM (IBM), American Airlines (AA), Delta Airlines (DAL), United Airlines (UAL), Pepsi (PEP), and Coca-Cola (KO).

```python
start = datetime.datetime(2013, 1, 1)   # histroy data load range
end = datetime.date.today()
```

```python
   def get(tickers, start, end):  #function to consolidate all the ticker into one DF
       def data(ticker):
           return web.DataReader(ticker, 'yahoo', start, end)
       datas = map(data, tickers)
       return pd.concat(datas, keys=tickers, names=['Ticker','Date'])
```

```python
tickers = ['AAPL','MSFT','GOOG','GE','IBM','AA','DAL','UAL', 'PEP', 'KO']
dfa = get(tickers, start, end)
```
