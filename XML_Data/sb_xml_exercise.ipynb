10 cities with the largest population

```python
import pandas as pd
cities=[]
year=[]
population=[]
for element in root.iterfind('country'):    
    for subelement in element.getiterator('city'):
        for pop in subelement.getiterator('population'):
            cities.append(subelement.find('name').text)
            year.append(pop.get('year'))
            population.append(int(pop.text))
dfp=pd.DataFrame(list(zip(cities,year,population)),columns=['cities','year','population'])
```

```python
dfg=dfp.groupby('cities').sum()  # grouby by city for all year and sum the population
```

10 ethnic groups with the largest overall populations (sum of best/latest estimates over all countries)

```python
ethnicg=[]
population=[]
for element in root.iterfind('country'):    
    for pop in subelement.getiterator('population'):
        if (element.find('ethnicgroup') != None):
            pop=int(pop.text)
            for eg in element.getiterator('ethnicgroup'):
                ethnicg.append(eg.text)
                population.append(pop * float(eg.get('percentage')))
dfeg=pd.DataFrame(list(zip(ethnicg,population)),columns=['ethnicgroup','population'])
```

```python
dfegroup=dfeg.groupby(ethnicg).sum()
```

```python
dfegroup.sort_values('population', ascending=False)[:10]
```

name and country of a) longest river, b) largest lake and c) airport at highest elevation

```python
# Longest RIVER
countries=[]
river=[]
length=[]
for element in root.iterfind('river'):
    if element.find('length')!= None:
        countries.append(element.get('country'))
        river.append(element.find('name').text)
        length.append(float(element.find('length').text))
dflr=pd.DataFrame(list(zip(countries,river,length)),columns=['countries','river','length'])
```

```python
dflr.sort_values('length',ascending=False)[:1]
```

```python

#largest lake
countries=[]
lake=[]
area=[]
for element in root.iterfind('lake'):
    if element.find('area')!= None:
        countries.append(element.get('country'))
        lake.append(element.find('name').text)
        area.append(float(element.find('area').text))
dfll=pd.DataFrame(list(zip(countries,lake,area)),columns=['countries','lake','area'])
```

```python
dfll.sort_values('area', ascending=False)[:1]
```
